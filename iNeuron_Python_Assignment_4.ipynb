### Q17. How do you distinguish between copy.copy() and copy.deepcopy()?

image.png

```python
import copy

original_list = [1, 2, [3, 4]]
copied_list = copy.copy(original_list)

# Modifying the nested list
copied_list[2].append(5)

print(original_list)
print(copied_list)


# We can see that as we are making changes in the original nested list, there is similar changes in copy
```

image.png

```python
import copy

original_list = [1, 2, [3, 4]]
deep_copied_list = copy.deepcopy(original_list)

# Modifying the nested list
deep_copied_list[2].append(5)

print(original_list)        # Output: [1, 2, [3, 4]]
print(deep_copied_list)     # Output: [1, 2, [3, 4, 5]]


# Although we are changing the nested list in original there is no change in the deepcopy
```
