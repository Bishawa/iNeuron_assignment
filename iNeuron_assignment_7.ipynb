```python
# Create two int type variables
first_variable = 10
second_variable = 5

# Perform the operations and store the results in variables
addition_result = first_variable + second_variable
subtraction_result = first_variable - second_variable
multiplication_result = first_variable * second_variable
division_result = first_variable / second_variable

# Print the results in the specified format
print("First variable is", first_variable, "& second variable is", second_variable)
print("Addition:", first_variable, "+", second_variable, "=", addition_result)
print("Subtraction:", first_variable, "-", second_variable, "=", subtraction_result)
print("Multiplication:", first_variable, "*", second_variable, "=", multiplication_result)
print("Division:", first_variable, "/", second_variable, "=", division_result)
```

```python
# Difference between '/' and '//'

a = 10
b = 3

result_normal_division = a / b
result_floor_division = a // b

print("Normal division:", result_normal_division)  
print("Floor division:", result_floor_division)    
```

```python
# Difference between '**' and '^'
base = 2
exponent = 3

result_exponentiation = base ** exponent

print("Exponentiation:", result_exponentiation)  # Output: 8

num1 = 2  # Binary: 101
print('binary of num 1 is: ', bin(num1))
num2 = 3  # Binary: 011
print('binary of num 2 is: ', bin(num2))

result_bitwise_xor = num1 ^ num2

print("Bitwise XOR:", result_bitwise_xor)  
```

### Q.4. Explain right shift operator and left shift operator with examples.



```python
number = 16    
print(bin(number))
shift_amount = 2
result = number >> shift_amount

print(f'{number} after right_shift is {result} which is because of binary code shifting to two positions to right')  
print(bin(result)) 
```

```python
number = 16    
print(bin(number))
shift_amount = 2
result = number << shift_amount

print(f'{number} after left_shift is {result} which is because of binary code shifting to two positions to left')  
print(bin(result)) 
```
