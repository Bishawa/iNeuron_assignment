### Q5 If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.keys()?



```python
spam = {'cat': 'meow', 'dog': 'bark'}

if 'cat' in spam:
    print('hello cat')
else:
    print('cat is not a spam')
```

```python
if 'cat' in spam.keys():
    print('hello cat')
else:
    print('cat is not a spam')
```

### Q6 If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values()?



```python
spam = {'meow': 'cat', 'bark': 'dog'}

if 'cat' in spam:
    print('hello cat')
else:
    print('cat is not a spam')
```

```python
if 'cat' in spam.values():
    print('hello cat')
else:
    print('cat is not a spam')
```

### What is a shortcut for the following code?
if 'color' not in spam:
spam['color'] = 'black'



```python
spam = {'meow': 'cat', 'bark': 'dog'}

spam.setdefault('color', 'black')

print(spam)
```

### Q7. How do you "pretty print" dictionary values using which module and function?



```python
from pprint import pprint

## defining the dictionary for pprint

spam = {'meow': 'cat', 'bark': 'dog', 'color': 'black'}

## Giving command for pprint
pprint(spam)
```
