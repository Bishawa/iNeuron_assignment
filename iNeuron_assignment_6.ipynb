### Q8. Explain the mutable and immutable data types with proper explanation & examples.



```python
# List is a mutable data type
my_list = [1, 2, 3, 4, 5]
print(my_list)  


# Modifying the list
my_list[2] = 10
print(my_list)  
# We can change the value at index 2 from 3 to 10
```

```python
# Strings are immutable data types
my_string = "Hello"
print(my_string)  

# Trying to modify the string (which is not possible)
# This will raise an error: TypeError: 'str' object does not support item assignment
my_string[0] = 'M'
```
