### Q9. How do you make a function variable refer to the global variable?
image.png

```python
# Example of local and global variable
x = 10
def func1():
  x = 20
  return x
```

```python
# giving the global variable
print(x)
```

```python
# Giving the local variable
print(func1())
```

```python
# Making a function variable refer to the global variable

x = 10

def func2():
  global x   # Declaring x as a globale function within the variable
  x = 20
```

```python
# Before calling the function
print(x)
```

```python
# After calling the function
func2()
print(x)
```

### Q13. What can you do to save a programme from crashing if it encounters an error?
image.png

```python
# Example for try and except error
# We want to create a program to divide two numbers
# we want to handle error which occurs i) if there input is string  ii) if denominator is zero
while True:
  try:
    a = int(input('Enter 1st number'))
    b = int(input('Enter 2nd number'))
    c = a/b
    print(c)
    break
  except ValueError:
    print('Please input an interger, strings are not accepted')
  except ZeroDivisionError:
    print('Please input a non-zero number')

```
