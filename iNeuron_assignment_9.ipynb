### 1. What is a lambda function in Python, and how does it differ from a regular function?





```python
# defining a lambda function to add two numbers
l_func = lambda a, b: a + b
l_func(56, 34)
```

```python
# Defining a regular function to add two numbers
def r_func(x, y):
    return x + y

r_func(56, 34)
```

### 2. Can a lambda function in Python have multiple arguments? If yes, how can you define and use them?



```python
# Example to show that lambda function can have multiple arguments
l1 = [34, 56, 78]
l2 = [23, 45, 34]
l3 = [56, 78, 65]


multiply = lambda x, y, z: x * y * z

result = list(map(multiply, l1, l2, l3))
print(result)
```

### 11. Write a recursive function to calculate the factorial of a given positive integer.

```python
# Recursive function to calculate the factorial of a given positive integer
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(7)
```

### 12. Implement a recursive function to compute the nth Fibonacci number.



```python
def fibonacci(n):
    if n <= 0:
        return None
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
fibonacci(12)
```

```python
fibonacci(4)
```

```python
fibonacci(5)
```

### 13. Create a recursive function to find the sum of all the elements in a given list.

```python
# Recursive function to find the sum of all the elements in a given list

def list_sum(lst):
    if not lst:  # Base case: If the list is empty, return 0
        return 0
    else:
        return lst[0] + list_sum(lst[1:])  # Recursive case: Add the first element to the sum of the rest of the list
```

```python
list_sum([34, 56, 78, 23, 45, 67])
```

### 14. Write a recursive function to determine whether a given string is a palindrome.



```python
def is_palindrome(s):
    # Convert the string to lowercase and remove spaces and punctuation
    s = ''.join(c.lower() for c in s if c.isalnum())

    if len(s) <= 1:
        return True
    else:
        if s[0] == s[-1]:
            return is_palindrome(s[1:-1])
        else:
            return False
```

### 15. Implement a recursive function to find the greatest common divisor (GCD) of two positive integers.

```python
# Writing a recursive function to find the GCD

def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
```

```python
gcd(48, 18)
```
